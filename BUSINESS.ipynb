# Choose the Right Structure

A Delaware S-corp can be transitioned into a Delaware C-corp, which is 
essentially the standard structure for publicly traded companies. Even 
though a publicly traded company may not be the eventual goal, it is a 
great path to have.

# Get Insurance

Virtually all busineses have some form of liability. Trade associations in a 
lot of industries offer cheaper insurance.

# Wait to Hire

Wait as long as possible before bringing on any employees.

# Set Goals

Set both personal and professional goals. 

It's easy to reverse engineer a problem given a goal. Keep the goals in front of you as much as possible, never lose sight of them.

Set a reasonable goal. Then double it... and then double it again.

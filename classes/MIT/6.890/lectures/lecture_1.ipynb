## A Crash Course on Complexity

### Polynomial Time

$P$ = Problems that can be solved in $n^O(i)$ time.

### Exponential time

$EXP$ = Problems that can be solved in $n^i$ time.

### Finite time

And $F$ - can we just call this $i$ time? 

Polynomial vs. Exponential time is the most common question. 

### Non-Deterministic Polynomial Time

$NP$ = Decision problems that are solvable via a "*lucky*" algorithm. Let's say that it just guesses, and then says yes or no.. but guesses are garanteed to always lead to a YES outcome if possible. CoNP is the opposite - just flip yes and no.

Another way to think of it...

$NP$ = Decision problems with solutions that can be "checked" in polynomial time. 

Lucky algorithms can be converted into checking algorithms and vice versa.

### P=NP?

When you say $P\neq NP$ you are essentially saying that luck cannot be engineered.

Personally, I'm not sure that's true. I think $P=NP$ might be true.

### Terminology

**X-Hard** = "At least as hard as every problem $\in X$"

**X-Complete** = Just the logical AND of two things... being X-hard and being in X. So, an X-Complete problem is X-Hard $\land \in X$

**PSPACE** = Problems solvable in polynomial space. Generally, problems that are solvable in PSPACE are solvable in EXP time because there are only EXP-many states in the machine.
